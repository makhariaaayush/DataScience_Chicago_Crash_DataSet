# 1. The data set need cleaning. Decide what to do with missing values and extra attributes.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Reading the csv file and storing it in df
df = pd.read_csv('/Users/aayushmakharia/Desktop/P3/Traffic_Crashes_-_Crashes.csv')
df.head(10)
```

```python
# finding the percentage of missing data in each column
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, (pct_missing*100)))
```

```python
# Replacing Missing value with NaN
df.replace('',np.nan)
```

```python
# Changing values to make it more compatible to generate appropriate visualization
print("Before Replace")
print(df['INTERSECTION_RELATED_I'].head(10))
df['INTERSECTION_RELATED_I'] = df['INTERSECTION_RELATED_I'].apply(lambda x : '1' if x=='Y'else('0' if x=='N' else x))
print("After Replace")
print(df['INTERSECTION_RELATED_I'].head(10))
```

```python
# Changing values to make it more compatible to generate appropriate visualization
print("Before Replace")
print(df['NOT_RIGHT_OF_WAY'].head(10))
df['NOT_RIGHT_OF_WAY'] = df['NOT_RIGHT_OF_WAY'].apply(lambda x : '1' if x=='Y'else('0' if x=='N' else x))
print("After Replace")
print(df['NOT_RIGHT_OF_WAY'].head(10))
```

```python
# Changing values to make it more compatible to generate appropriate visualization
print("Before Replace")
print(df['HIT_AND_RUN_I'].head(10))
df['HIT_AND_RUN_I'] = df['HIT_AND_RUN_I'].apply(lambda x : '1' if x=='Y'else('0' if x=='N' else x))
print("After Replace")
print(df['HIT_AND_RUN_I'].head(10))
```

```python
# Changing values to make it more compatible to generate appropriate visualization

print("Before Replace")
print(df['CRASH_DAY_OF_WEEK'].head(10))
df['CRASH_DAY_OF_WEEK'] = df['CRASH_DAY_OF_WEEK'].apply(lambda x : 'Monday' if x==2
                                                        else('Tuesday' if x==3
                                                             else('Wednesday' if x==4
                                                                  else('Thursday' if x==5
                                                                       else('Friday' if x==6
                                                                            else('Saturday' if x==7
                                                                                 else('Sunday' if x==1 else x)))))))
print("After Replace")
print(df['CRASH_DAY_OF_WEEK'].head(10))
```

```python
# Changing values to make it more compatible to generate appropriate visualization
print("Before Replace")
print(df['CRASH_MONTH'].head(10))
df['CRASH_MONTH'] = df['CRASH_MONTH'].apply(lambda x : 'January' if x==1
                                                  else('February' if x==2
                                                  else('March' if x==3 
                                                  else('April' if x==4
                                                  else('May' if x==5
                                                  else('June' if x==6
                                                  else('July' if x==7 
                                                  else('August' if x==8      
                                                  else('September' if x==9   
                                                  else('October' if x==10 
                                                  else('November' if x==11
                                                  else('December' if x==12 else x))))))))))))
print("After Replace")
print(df['CRASH_MONTH'].head(15))
```

```python
# Changing values to make it more compatible to generate appropriate visualization
df['TRAFFIC_CONTROL_DEVICE'] = df['TRAFFIC_CONTROL_DEVICE'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else x)
df['DEVICE_CONDITION'] = df['DEVICE_CONDITION'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else x)
df['WEATHER_CONDITION'] = df['WEATHER_CONDITION'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else x)
df['LIGHTING_CONDITION'] = df['LIGHTING_CONDITION'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else x)
df['TRAFFICWAY_TYPE'] = df['TRAFFICWAY_TYPE'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else('OTHER' if x=='UNKNOWN INTERSECTION' else x))
df['ROADWAY_SURFACE_COND'] = df['ROADWAY_SURFACE_COND'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else x)
df['ROAD_DEFECT'] = df['ROAD_DEFECT'].apply(lambda x : 'OTHER' if x=='UNKNOWN' else x)
df['DAMAGE'] = df['DAMAGE'].apply(lambda x : '1' if x=='OVER $1,500'
                                                  else('2' if x=='$501 - $1,500'
                                                  else('3' if x== '$500 OR LESS' else x)))
```

# 2. Some attributes are more useful if you break them into several attributes. An example of this is already included in the data set where the time, day, and month of the crash are given as separate attributes. These attributes allow you to compare crashes based on the day of the week, time, or month (season). Are there other attributes that you can break down into smaller attributes to gain more information from?

```python
# Dividing the attribute into smaller attributes
df['DATE_POLICE_NOTIFIED']  = pd.to_datetime(df['DATE_POLICE_NOTIFIED'])

df['DATE_POLICE'] = df['DATE_POLICE_NOTIFIED'].dt.date
df['TIME_POLICE'] = df['DATE_POLICE_NOTIFIED'].dt.time
df['HOUR_POLICE'] = df['DATE_POLICE_NOTIFIED'].dt.hour
df['MINUTE_POLICE'] = df['DATE_POLICE_NOTIFIED'].dt.minute
df['CRASH_DATE']  = pd.to_datetime(df['CRASH_DATE'])
df['CRASH_YEAR'] = df['CRASH_DATE'].dt.year
```

```python
# Displaying the altered datafram df
df
```

```python
# Data Preparation
df['CRASH_SEASON'] = df['CRASH_MONTH'].apply(lambda x : 'Winter' if x=='January'
                                                  else('Winter' if x=='February'
                                                  else('Spring' if x=='March'
                                                  else('Spring' if x=='April'
                                                  else('Spring' if x=='May'
                                                  else('Summer' if x=='June'
                                                  else('Summer' if x=='July'
                                                  else('Summer' if x=='August'     
                                                  else('Fall' if x=='September' 
                                                  else('Fall' if x=='October'
                                                  else('Fall' if x=='November'
                                                  else('Winter' if x=='December' else x))))))))))))
```

```python
# Data Preparation
df['POSTED_SPEED_LIMIT'] = df['POSTED_SPEED_LIMIT'].apply(lambda x : '0-10' if x> -1 and x< 11
                                                  else('11-20' if x> 10 and x< 21
                                                  else('21-30' if x> 20 and x< 31
                                                  else('31-40' if x> 30 and x< 41
                                                  else('41-50' if x> 40 and x< 51
                                                  else('51-60' if x> 50 and x< 61
                                                  else('61-70' if x> 60 and x< 71
                                                  else('71-80' if x> 70 and x< 81
                                                  else('81-90' if x> 80 and x< 91
                                                  else('91-100' if x> 90 and x< 101 else x))))))))))
```

```python
# Data Preparation
df['DAY_OR_NIGHT'] = df['CRASH_HOUR'].apply(lambda x : 'Night' if (x > 19 or x < 6) else 'Day' )
```

# 3. What are some insights about the crashes and date/time? You can look into season, day of the week, day/night, lightning, weather, etc

```python
# 3.1
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 5, 5
x = df.groupby("CRASH_SEASON").size()
plt.ylabel("Count")
plt.title("No. of Crashes")
x.plot.bar()
x
```

```python
# 3.2
x1 = df.groupby("WEATHER_CONDITION").size()
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
plot = x1.plot(x='Weather', y='Count', xticks=range(0,12),fontsize=9)
x1.plot.line()
x1
```

```python
# 3.3
x2 = df.groupby("CRASH_DAY_OF_WEEK").size().sort_values(ascending=True)
from pylab import rcParams
rcParams['figure.figsize'] = 8, 8
plt.title("Day-To-Day Crash")
x2.plot.pie(autopct='%1.1f%%')
x2
plt.ylabel('')
```

```python
# 3.4
x3 = df.groupby("CRASH_MONTH").size().sort_values(ascending=False)
plot = x3.plot(x='Month', y='Count', xticks=range(0,12))
rcParams['figure.figsize'] = 25,20
x3.plot.line(marker="o")
x3.plot()
x3
```

```python
# 3.5
x4 = df.groupby("DAY_OR_NIGHT").size().sort_values(ascending=False)
plt.ylabel("Count")
plt.title("Day/Night")
from pylab import rcParams
rcParams['figure.figsize'] = 8, 8
x4.plot.bar()
x4
```
